**Part B:** How do the RIGHT-joined tables differ from the LEFT-joined tables in Question 1?  

> Your Answer: Left Join keeps all the left values and will place a null value with non matching from the right; Right join is vice versa

> Your Answer:  Update the query in the code box below.

```sql

-- INNER JOIN
SELECT DISTINCT b.title, b.average_rating, b.books_count, tr.user_id, 

tr.book_id
FROM BooksDB.dbo.books AS b
INNER JOIN BooksDB.dbo.to_read AS tr
ON b.book_id = tr.book_id
WHERE b.title IS NOT NULL
ORDER BY tr.book_id;
```
